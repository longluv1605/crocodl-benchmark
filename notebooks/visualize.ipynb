```python
import numpy as np
from scipy.spatial.transform import Rotation as R

import sys
import os

# Add project root to path
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.insert(0, project_root)
    
from visualize.cam_pose_visualizer import CamPoseVisualizer, load_gt_and_est_poses

visualizer = CamPoseVisualizer()
```

```python
ALM_TRAJ_COLOR = [255, 0, 0]        # Red
QUERY_TRAJ_COLOR = [0, 255, 0]      # Green
AUG_TRAJ_COLOR = [0, 0, 255]        # Blue
MAP_TRAJ_COLOR = [255, 255, 255]    # White

def visualize_device(device='ios'):
    query_traj_path = f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_query/trajectories.txt'
    alm_traj_path = f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_query/proc/alignment_trajectories.txt'
    sensors_path = f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_query/sensors.txt'
    rigs_path = None if device == 'ios' else f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_query/rigs.txt'
    alm_traj_poses, query_traj_poses = load_gt_and_est_poses(alm_traj_path, query_traj_path, sensors_path, rigs_path, ALM_TRAJ_COLOR, QUERY_TRAJ_COLOR)
    
    map_traj_path = f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_map/trajectories.txt'
    aug_traj_path = f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_map/trajectories_augumented.txt'
    sensors_path = f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_map/sensors.txt'
    rigs_path = None if device == 'ios' else f'/home/long/Workspace/crocodl-benchmark/capture/ARCHE_D2/sessions/{device}_map/rigs.txt'
    aug_traj_poses, map_traj_poses = load_gt_and_est_poses(aug_traj_path, map_traj_path, sensors_path, rigs_path, AUG_TRAJ_COLOR, MAP_TRAJ_COLOR)
    
    visualizer.visualize(map_traj_poses + query_traj_poses + alm_traj_poses + aug_traj_poses)
    return map_traj_poses, query_traj_poses, alm_traj_poses, aug_traj_poses
```
